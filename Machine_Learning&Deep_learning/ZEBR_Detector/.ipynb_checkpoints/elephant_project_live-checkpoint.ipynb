```python
import torch
import cv2
from ultralytics import YOLO

# Check PyTorch and CUDA setup
print("PyTorch Version:", torch.__version__)
print("CUDA Version:", torch.version.cuda)
print("Is CUDA available?", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU Device Name:", torch.cuda.get_device_name(0))
```

```python


# Load the model
model = YOLO('best.pt')  # Load the trained model
model.to('cuda')  # Move the model to GPU

# Open a connection to the camera
cap = cv2.VideoCapture(0)  # Change the index if you have multiple cameras
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference
    results = model(frame)

    # Loop through results and draw bounding boxes
    for result in results:
        boxes = result.boxes  # Extract the bounding boxes
        for box in boxes:
            conf = box.conf.cpu().item()  # Confidence score

            if conf > 0.80:  # Check if confidence is greater than 0.80
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()  # Bounding box coordinates
                cls = int(box.cls.cpu().item())  # Class ID

                # Get class label
                label = f'{model.names[cls]}: {conf:.2f}'

                # Draw the bounding box and label on the frame
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Display the frame
    cv2.imshow('YOLO Live Detection', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cap.release()
cv2.destroyAllWindows()
```
